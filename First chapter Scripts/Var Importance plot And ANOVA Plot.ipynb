```python
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
import os
import seaborn as sns
sns.set()

os.chdir('D:/Pesquisa/Mestrado/PRIMEIRO_CAPITULO_DIFERENCIADO/')
os.listdir()
dataSen = pd.read_csv('PlanilhaoABOVASENSIBILIDADE_PELOTSS_v2.txt',sep = "\t")
dataTSS = pd.read_csv('PlanilhaoANOVATSS_PELOTSS_v2.txt',sep = "\t")
dataAUC = pd.read_csv('PlanilhaoAUC.txt',sep = "\t")
dataTypeII = pd.read_csv('TypeIIError.txt',sep ="\t" )
print(dataSen.columns)
print(dataTSS.columns)
print(dataAUC.columns)
print(dataTypeII.columns)
```

```python
#Botando os valores em ordem alfabética:
dataSen=dataSen.sort_values('Data_Partitioning',ignore_index = True)
print(dataSen)
dataTSS=dataTSS.sort_values('Data_Partitioning',ignore_index = True)
print(dataTSS)
dataAUC=dataAUC.sort_values('Modelo',ignore_index = True)
print(dataAUC)
dataTypeII=dataTypeII.sort_values('Data_Partitioning',ignore_index = True)
print(dataTypeII)
```

```python
novo_nome = {'FullSet':'FullSet','RdPd':"RdUd",'PdRd':"UdRd"}
dataTSS['Data_Partitioning'] = dataTSS['Data_Partitioning'].map(novo_nome)
dataSen['Data_Partitioning'] = dataSen['Data_Partitioning'].map(novo_nome)
dataAUC['Modelo'] = dataAUC['Modelo'].map(novo_nome)
dataTypeII['Data_Partitioning'] = dataTypeII['Data_Partitioning'].map(novo_nome)
dataTSS.head(-30)
```

```python
fig, axes = plt.subplots(2, 2, sharey=False, figsize=(25,15))



pal = sns.color_palette('Blues_r',3)
sns.lineplot(ax = axes[0,0],x=dataSen['Algoritmo'], y=dataSen['Sensibilidade'],hue = dataSen['Data_Partitioning'],ci=None,style =dataSen['Data_Partitioning'],markers=("s","o","D"),
            dashes=[(1,1), (3, 3),(5,5)],palette=pal, legend = False,ms = 12)

sns.lineplot(ax = axes[0,1],x=dataAUC['Algoritmo'], y=dataAUC['AUC'],hue = dataAUC['Modelo'],ci=None,style =dataAUC['Modelo'],markers=("s","o","D"),
            dashes=[(1, 1), (3, 3),(5,5)],palette=pal, legend = False,ms = 12)

sns.lineplot(ax = axes[1,0],x=dataTypeII['Algoritmo'], y=dataTypeII['Type_II_Error'],hue = dataTypeII['Data_Partitioning'],ci=None,style =dataTypeII['Data_Partitioning'],markers=("s","o",'D'),
            dashes=[(1, 1), (3, 3),(5,5)],palette=pal, legend = False,ms = 12)


lp =sns.lineplot(ax = axes[1,1],x=dataTSS['algoritmos'], y=dataTSS['TSS'],hue = dataTSS['Data_Partitioning'],ci=None,style =dataTSS['Data_Partitioning'],markers=("s","o",'D'),
            dashes=[(1, 1), (3, 3),(5,5)],palette=pal,ms = 12)



lp.legend(loc='upper right', bbox_to_anchor=(1.30, 2.2),fontsize = 25,facecolor='white')


axes[0,0].set_xlabel(None)
axes[0,0].set_ylabel('Sensibility',fontsize = 25)
axes[0,0].grid(False)
axes[0,0].set_facecolor('white')
axes[0,0].tick_params(axis='both', which='major', labelsize=20)

axes[0,1].set_xlabel(None)
axes[0,1].set_ylabel('AUC',fontsize = 25)
axes[0,1].grid(False)
axes[0,1].set_facecolor('white')
axes[0,1].tick_params(axis='both', which='major', labelsize=20)

axes[1,0].set_xlabel(None)
axes[1,0].set_ylabel('Type II Error',fontsize = 25)
axes[1,0].grid(False)
axes[1,0].set_facecolor('white')
axes[1,0].tick_params(axis='both', which='major', labelsize=20)

axes[1,1].set_xlabel(None)
axes[1,1].set_ylabel('TSS',fontsize = 25)
axes[1,1].grid(False)
axes[1,1].set_facecolor('white')
axes[1,1].tick_params(axis='both', which='major', labelsize=20)
```
